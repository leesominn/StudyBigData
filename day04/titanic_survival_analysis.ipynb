#### 기본 패키지 import

```python
import seaborn as sns
import pandas as pd 
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.to_csv('./data/titanic.csv', index=True, encoding='utf-8')
```

#### 결측치(NaN) 파악

```python
titanic.isnull().sum()
```

##### 결측치 치환

- 891명의 데이터에서 중앙값을 찾아 결측치를 치환

```python
titanic.describe()
```

```python
# age 결측치 채우기
median = titanic.age.median()
titanic['age'] = titanic['age'].fillna(median) 
```

```python
titanic.isnull().sum()
```

```python
# embarked(=embark_town : 승선장소) 값 확인
titanic.embarked.value_counts()
```

```python
# embarked, embark_town 결측치 채우기
titanic.embarked = titanic.embarked.fillna('S')
titanic.embark_town = titanic.embark_town.fillna('Southampton')
```

```python
titanic.isnull().sum()
```

```python
# deck 값 확인
titanic.deck.value_counts()
```

```python
# deck 결측치 채우기
titanic.deck = titanic.deck.fillna('C')
```

```python
titanic.isnull().sum()
```

#### titanic 데이터 기본정보 확인

```python
titanic.info()
```

```python
titanic.survived.value_counts()
```

#### 차트 시각화

```python
import matplotlib.pyplot as plt
```

##### 성별에 따른 생존율 차트

```python
f, ax = plt.subplots(1,2, figsize=(10,5))
titanic.survived[titanic.sex == 'male'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0], explode=[0,0.1], shadow=True)
# ax:표 위치, explode:차트가 떼어지는 것, shadow=차트에 그림자
titanic.survived[titanic.sex == 'female'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[1], explode=[0,0.1], shadow=True)
ax[0].set_title('Survived (Male)')
ax[1].set_title('Survived (Female)')
plt.show()
```

##### 객실등급에 따른 생존율 차트

```python
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Pclass vs Survived')
plt.show()
```

##### 상관계수 확인

#### 상관분석 결과 시각화

```python
# sns.set(style='ticks', color_codes=True)
# adult_male, alone은 bool타입은 pairplot(산점도)를 그릴 때 오류를 낼 수 있기 때문에 int32로 타입을 변환시켜야함
titanic = titanic.astype({'adult_male': 'int32', 'alone': 'int32'})

sns.pairplot(titanic, hue='survived')
plt.show()
```
