```python
all_data = [] # 데이터 저장할 리스트 만들기
for file in all_files:
    df = pd.read_excel(file)
    all_data.append(df)

all_data[0]
```

```python
all_data_concat = pd.concat(all_data, axis=0, ignore_index=True) # 데이터 병합하기
all_data_concat = all_data_concat.drop(['Unnamed: 0', '번호'], axis='columns')
all_data_concat
```

```python
all_data_concat.to_csv('./data/riss_bigdata.csv', index=False, encoding='utf-8')
```

#### 데이터 전처리

```python
all_title = all_data_concat['제목']
all_title
```

```python
stopWords = set(stopwords.words('english')) # 영어 불용어를 불러와서 저장하기
stopWords
```

```python
lemma = WordNetLemmatizer() # 표제어 추출 작업을 제공하는 함수
```

```python
words = []

for title in all_title:
    EnWords = re.sub(r'[^a-zA-z]+', ' ', str(title)) # 알파벳으로 시작하지 않는 단어는 공백으로 치환하여 제거(re.sub)
    EnWordsToken = word_tokenize(EnWords.lower()) # 단어들을 소문자로 정규화
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords] # 단어 토큰화
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop] # 불용어(ex. a, the, and) 제거
    
    words.append(EnWordsTokenStopLemma)
```

```python
print(words) # 불용어 제거로 키워드 단어만 남음
```

```python
# 2차원리스트(제목별로 단어가 리스트로) -> 1차원리스트로(단어들이 전체 리스트로)
words_reduce = list(reduce(lambda x,y : x+y, words))
print(words_reduce)
```

```python
count = Counter(words_reduce)
count
```

```python
# 출현 횟수가 많은 상위 30개 단어 중 단어 길이가 1보다 큰 것만 word_count 딕셔너리에 저장하고 출력하기
word_count = dict()

for tag, counts in count.most_common(30):
    if len(str(tag)) > 1:
        word_count[tag] = counts
        print(f'{tag} : {counts}')
```

#### 단어 개수 차트

```python
plt.figure(figsize=(20,5))
sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='90')
plt.show()
```

#### 결과 시각화

```python
all_data_concat['doc_count']= 0
summary_year = all_data_concat.groupby('출판일', as_index=False)['doc_count'].count()
summary_year
```

```python
# 그래프 그리기
plt.figure(figsize=(12,5))
plt.xlabel('year')
plt.ylabel('doc_count')
plt.grid(True)
plt.plot(range(len(summary_year)), summary_year['doc_count'])
plt.xticks(range(len(summary_year)), [text for text in summary_year['출판일']])
plt.show()
```

```python
# 워드클라우드 그리기
stopwords = set(STOPWORDS)
wc = WordCloud(background_color='ivory', stopwords=stopwords, width=1024, height=700)
cloud = wc.generate_from_frequencies(word_count)

plt.figure(figsize=(15,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
