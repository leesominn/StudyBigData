```python
import pandas as pd 
import folium
```

```python
cb_data = pd.read_csv('./data/CoffeeBean.csv', index_col=0, header=0, encoding='utf-8')
cb_data.head(20)
```

```python
# 현재 주소체계 확인
addr = []

for address in cb_data.address:
    addr.append(str(address).split())

addr
```

```python
addr2 = []

for i in range(len(addr)):
    if addr[i][0] == '서울': addr[i][0] = '서울특별시'
    if addr[i][0] == '서울시': addr[i][0] = '서울특별시'
    if addr[i][0] == '부산시': addr[i][0] = '부산광역시'
    if addr[i][0] == '인천': addr[i][0] = '인천광역시'
    if addr[i][0] == '광주': addr[i][0] = '광주광역시'
    if addr[i][0] == '대전시': addr[i][0] = '대전광역시'
    if addr[i][0] == '울산시': addr[i][0] = '울산광역시'
    if addr[i][0] == '세종시': addr[i][0] = '세종특별자치시'
    if addr[i][0] == '경기': addr[i][0] = '경기도'
    if addr[i][0] == '충북': addr[i][0] = '충청북도'
    if addr[i][0] == '충남': addr[i][0] = '충청남도'
    if addr[i][0] == '전북': addr[i][0] = '전라북도'
    if addr[i][0] == '전남': addr[i][0] = '전라남도'
    if addr[i][0] == '경북': addr[i][0] = '경상북도'
    if addr[i][0] == '경남': addr[i][0] = '경상남도'
    if addr[i][0] == '제주': addr[i][0] = '제주특별자치도'
    if addr[i][0] == '제주도': addr[i][0] = '제주특별자치도'
    if addr[i][0] == '제주시': addr[i][0] = '제주특별자치도'

    addr2.append(' '.join(addr[i]))

addr # 확인해보기 위한 출력
# print(len(addr2))
```

```python
addr2 = pd.DataFrame(addr2, columns=['address2'])
```

```python
# cb_data와 addr2를 열 방향(옆으로 연결)으로 결합하기
cb_data2 = pd.concat([cb_data, addr2], axis=1)
cb_data2.head()
```

```python
# cb_data2를 csv로 저장하기
cb_data2.to_csv('./data/CoffeeBean_2.csv', encoding='utf-8', index=False)
```

### folium 사용법

```python
map = folium.Map(location=[35.172801, 129.130621], zoom_start=15)
folium.Marker(location=[35.172801, 129.130621], popup='부산정보산업진흥원', icon=folium.Icon(color='blue', icon='glyphicon-map-marker')).add_to(map)
map
```

### 커피빈 매장주소 위/경도 이용해 마커 찍기

```python
cb_geoData = pd.read_csv('./data/CoffeeBean_geo.csv', encoding='utf-8')
cb_geoData.head()
```

```python
map_cb = folium.Map(location=[35.172801, 129.130621], zoom_start=15)
map_cb
```

```python

for i, store in cb_geoData.iterrows():
    # print(i)
    # print(store)
    # if i == 2: break
    folium.Marker(location=[store['위도'], store['경도']], popup=store['store'], icon=folium.Icon(color='purple', icon='home')).add_to(map_cb)

map_cb
```

```python
map_cb.save('./data/CoffeeBean_shop.html')
```
